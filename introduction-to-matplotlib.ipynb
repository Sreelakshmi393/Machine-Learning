# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width, height

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save & show (you save the wholw figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make some data for histograms and plot it
x = np.random.rand(10)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.plot(x,x/2);
ax1.set(title="This is ax1")
ax2.scatter(np.random.random(10), np.random.random(10));
ax2.set(title="This is ax 2")
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.set(title="This is ax 3")
ax4.hist(np.random.randn(1000));
ax4.set(title="This is ax 4");
```

```python
# Subplots options 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize =(10,5))
#Plot to each differnt index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.rand(1000));
```

# Which one should you use? (Pyplot Vs Matplolib OO Method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advance, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO Method with pyplot method

fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
# ax.set_xlim([45,100])
```

```python
over_50.target.values
```

```python
# OO Method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol,age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
       ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart disease and Thalach",
       xlabel="Age",
       ylabel="Max heart rate")

# Add a lengend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO Method from scratch

fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") # This changes the color scheme

# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Customizing x and y axis limitations

# Subplot of chol,age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
       ylabel="Cholesterol")

# Change the x axis limits
ax0.set_xlim([50,80])
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax1
ax1.set(title="Heart disease and Thalach",
       xlabel="Age",
       ylabel="Max heart rate")
# Change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a lengend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
